**파이썬**([Python](https://www.python.org))은 네덜란드 프로그래머인 [Guido van Rossum](https://en.wikipedia.org/wiki/Guido_van_Rossum)이 1989년에 고안한 범용 프로그래밍 언어이다. Python은 무료이고, 오픈 소스이며, [Python Software Foundation](https://www.python.org/psf/)을 통해 개발이 진행된다. Python은 가장 많이 사용되는 10대 프로그래밍 언어에 꾸준히 랭크 되었으며, 2024년 1월 [PYPL(PopularitY of Programming Language)](https://pypl.github.io/PYPL.html) 기준으로 전 세계적으로 가장 인기있는 프로그래밍 언어이다.(파이썬 다음으로 Java, JavaScript, C/C++, C#, R, PHP 순이다.) Python은 과학 컴퓨팅을 위한 주요 도구로 전세계적으로 과학 분야 작업에서 비중이 급속히 확대되고 있다.

Python은 우리가 여기에서 다룰 **데이터 분석**뿐만 아니라 웹 개발, CGI(common gateway interface) 및 그래픽 유저 인터페이스, 게임 개발, 멀티미디어, 데이터 처리, 보안 등 거의 모든 애플리케이션 도메인에서 사용되는 범용 언어이다. 파이썬은 Google, Dropbox, YouTube 등 인터넷 서비스 및 첨단테크 회사에서 광범위하게 사용되고 있다.

### Jupyter Notebook

**주피터 노트북**([Jupyter Notebook](http://jupyter.org/))은 파이썬 및 여러 과학 라이브러리와 상호 작용할 수 있는 방법 중 하나로서 **브라우저 기반** 인터페이스로 파이썬을 사용한다. 파이썬 명령을 작성하고 실행하며, 표, 그림, 애니메이션 등을 브라우저에 출력하고, 서식있는 텍스트와 수학 표현을 혼합하는 것이 가능하다. 이러한 기능으로 인해 주피터 노트북은 이제 과학 컴퓨팅 생태계의 메이저 플레이어가 되었다. 주피터 노트북은 아나콘다 배포판의 일부이다. 

Windows일 경우, 아나콘다를 설치하고 시작 버튼을 누르면, 아나콘다 앱이 새로 설치된 것을 확인할 수 있다. 그 안에 주피터 노트북이 새로 설치된 것을 확인할 수 있다.
  


#### 주피터 노트북을 여는 두 가지 방법

(1) 애플리케이션 메뉴에서 주피터 노트북을 선택하면 자신의 웹 브라우저에 주피터 노트북이 열린다. 이 경우 주피터 노트북이 실행되는 위치가 자신의 컴퓨터 사용자(Users) 디렉토리가 된다. 

(2) 주피터 노트북이 실행되는 특정 위치를 자신이 선택할 수도 있는데, 윈도즈의 경우 가령 C 드라이브의 루트 디렉토리(즉, `C:\`)에서 주피터 노트북이 실행되도록 하기 위해서는 다음과 같이 하면 된다. Windows의 경우, 우선 시작 버튼의 Anaconda 앱 중에서 아래와 같이 Jupyter Notebook 대신 Anaconda Prompt를 클릭한다.



아래와 같은 프롬프트 창이 나오는데 `(base) C:\Users\사용자이름>`에서 `cd\`를 타이핑 하고 엔터를 치고, 그 다음 `jupyter notebook`을 타이핑 하고 엔터를 치면 첫 번째 방법과 마찬가지로 자신의 웹 브라우저에 주피터 노트북이 열린다. 이 경우 사용자(Users) 디렉토리가 아니라 `C:\` 디렉토리에서 주피터 노트북이 구동된다.(이렇게 하면 주피터 노트북 파일을 넣어 놓은 디렉토리를 찾아가기가 쉽다.) 



이상의 모든 것이 정상적으로 작동한다고 가정하면 자신의 웹브라우저 상에 Jupyter Notebook이 다음과 같은 웹 페이지로 열렸을 것이다. 여기에서 오른쪽 상단의 `New`를 클릭하고 `Python 3`을 선택해보자.



위와 같이 했을 때 자신의 컴퓨터에 아래와 같은 페이지가 나타나며, Python 명령을 입력할 수있는 *active cell*이 표시된다.

### 숫자 프로그래밍

행렬(matrix) 및 배열(array)의 기본적 처리 기능은 넘파이([NumPy](http://www.numpy.org/)) 라이브러리에서 제공한다. 예를 들어, 다음과 같이 배열을 만들어보자. 배열이란 벡터, 행렬, 숫자표 등을 표현하기 위해 항목을 차례로 배열한 것을 말한다.

````python
import numpy as np                     # numpy 라이브러리 로드

a = np.linspace(-np.pi, np.pi, 100)    # -π에서 π까지 100개의 동일 간격 그리드 만들기
a
````

 

>**다음 파이썬 코드를 쉽게 설명해줘.**
>```
import numpy as np 
a = np.linspace(-np.pi, np.pi, 100) 
a
>```

---

>이 파이썬 코드는 `numpy` 라이브러리를 사용하여 특정한 수열을 생성합니다. 각 코드 라인의 의미는 다음과 같습니다.
>
>1. `import numpy as np`:
    - `numpy` 라이브러리를 `np`라는 약칭으로 가져옵니다. `numpy`는 수학적 계산을 위한 파이썬 라이브러리로, 특히 배열이나 행렬 연산에 자주 사용됩니다.
>
>2. `a = np.linspace(-np.pi, np.pi, 100)`:
    - `np.linspace()` 함수는 지정된 범위 내에서 균일한 간격의 숫자들을 생성합니다. 
    - 여기서는 `-np.pi`부터 `np.pi`까지의 범위에 걸쳐 총 100개의 숫자를 생성합니다. `-np.pi`와 `np.pi`는 각각 원주율 π의 음수와 양수 값을 나타냅니다.
    - 결과적으로, 이 라인은 -π부터 π까지의 범위에서 균등하게 분포된 100개의 숫자를 포함하는 배열 `a`를 생성합니다.
>
>3. `a`:
    - 이 라인은 배열 `a`를 출력합니다. 이 배열에는 -π에서 π까지 균등하게 분포된 100개의 숫자가 들어 있습니다.
>
>이 코드는 주로 수학이나 공학 계산, 데이터 시각화 등에서 사용됩니다. 예를 들어, 함수의 그래프를 그릴 때 x축의 값을 생성하는 데 사용할 수 있습니다. 여기서 생성된 값들은 원주율 π에 관련된 계산이나 trigonometric(삼각 함수) 함수의 그래프를 그릴 때 유용합니다.

#### 파이썬 라이브러리

**라이브러리**(library)는 기본 파이썬 배포판에는 포함되지 않은 모듈 또는 함수 그룹을 로딩하는 데 사용된다. 아주 기본적인 분석을 수행하는 함수는 기본 파이썬 배포판에 표준으로 제공되지만 좀 더 다양하고 복잡한 함수나 작업을 위해서는 해당 라이브러리를 불러와서 사용한다. 참고로 2022년 11월 현재 Python 소프트웨어의 공식 저장소인 PyPI(Python Package Index)에는 다양한 기능을 가진 41만개 이상의 라이브러리 패키지가 있다.(Wikipedia, "[Python (programming language)][key]") 

[key]: https://en.wikipedia.org/wiki/Python_(programming_language)

아나콘다 배포판을 설치하고 주피터 노트북을 통해 파이썬을 사용하는 경우, 머신러닝 통계 분석에 필요한 각종 표준적인 라이브러리 패키지를 사용자가 직접 다운로드할 필요가 없고, 이미 이런 라이브러리들이 설치돼있다. 단, 설치된 라이브러리를 사용하기 위해서는 그 때마다 바로 위 명령문의 예처럼 해당 라이브러리를 호출하는 절차(즉, `import numpy as np`)가 필요하다. 

[SciPy](http://www.scipy.org) 라이브러리는 NumPy 위에 구축되었으며, 예를 들어 $ \int_{-2}^2 \phi(z) dz $와 같은 적분 계산이 가능하다. 여기서 $ \phi $는 표준정규(standard normal) 확률밀도함수(probability density function)이다.

````python
from scipy.stats import norm
from scipy.integrate import quad

ϕ = norm()
value, error = quad(ϕ.pdf, -2, 2)  # 가우시안 구적법(Gaussian quadrature)을 사용하여 적분 실행
value
````

### Symbolic Algebra

Mathematica 또는 Maple에서처럼 기호 표현 계산 등이 가능하다. [SymPy](http://www.sympy.org/) 라이브러리는 이 기능을 제공한다.

````python
from sympy import Symbol

x, y = Symbol('x'), Symbol('y')  # 'x'와 'y'를 대수 기호로 취급
x + x + x + y
````

이 경우 다음과 같은 조작이 가능하다.

````python
expression = (x + y)**3
expression.expand()
````

또한 다항식 풀기가 가능하다.

### 통계 분석

Python의 데이터 관리 및 통계 라이브러리는 지난 몇 년 동안 빠르게 발전해 왔다. 데이터 작업에 가장 많이 사용되는 라이브러리 중 하나는 [pandas](http://pandas.pydata.org/)이다. 다음은 간단한 예이다.

````python
import pandas as pd
np.random.seed(1234)

data = np.random.randn(5, 2)  # N(0,1) 정규분포에서 5x2 행렬 임의 추출
dates = pd.date_range(start='1/1/2022', end='1/5/2022')

df = pd.DataFrame(data, columns=('price', 'weight'), index=dates)
print(df)
````

````python
# 바로 위에서 만든 'df'라는 데이터 프레임의 평균값 구하기
df.mean()
````

>**%pip install과 !pip install의 차이점을 알려줘.**

---

>Jupyter 노트북에서 `%pip install`과 `!pip install`은 Python 패키지를 설치하는 데 사용되지만, 둘 사이에는 약간의 차이가 있습니다.
>
>1. `!pip install`:
    - 이 구문은 Jupyter 노트북에서 시스템 셸 명령을 실행하기 위해 사용됩니다.
    - `!` 기호는 셸 명령을 실행하는 데 사용되며, 이 경우 `pip` 설치 명령을 실행합니다.
    - `!pip install`은 현재 Jupyter 노트북이 실행 중인 환경에 패키지를 설치합니다.
    - 하지만 `!pip install`은 Jupyter 노트북의 커널과 정확히 일치하지 않는 Python 환경에 패키지를 설치할 위험이 있습니다. 예를 들어, 다른 Python 버전이나 환경이 활성화되어 있을 때 문제가 발생할 수 있습니다.
>
>2. `%pip install`:
    - `%pip install`은 Jupyter 노트북에 특화된 명령어로, Jupyter 노트북의 매직 명령어 중 하나입니다.
    - 이 명령어는 Jupyter 노트북이 사용 중인 Python 커널과 연결된 환경에 패키지를 설치하는 데 사용됩니다.
    - `%pip install`은 사용 중인 커널과 일치하는 환경에 패키지를 설치하여, 커널과 환경 간의 불일치 문제를 방지합니다.
    - 이는 보다 안정적이고 예측 가능한 패키지 설치 방법을 제공합니다.
>
>요약하자면, `%pip install`은 Jupyter 노트북의 현재 커널 환경에 더욱 정확하게 패키지를 설치하는 반면, `!pip install`은 시스템 셸을 통해 패키지를 설치하지만, 환경 불일치의 위험이 있습니다. 따라서 Jupyter 노트북에서 패키지를 설치할 때는 `%pip install`을 사용하는 것이 더 권장됩니다.
